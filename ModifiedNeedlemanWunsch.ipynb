```python
sequences=['ACACTG','ACACTGATCG']
a=sequences[0]
b=sequences[1]
space=' '
a=space+a
b=space+b
```

```python
import numpy as np
import pandas as pd
m=len(a) #rows of a matrix
n=len(b) #columns of a matrix
matrix = np.zeros( (m,n))
name2 = [x for x in b]
name1 = [y for y in a]
df = pd.DataFrame(matrix, index=name1, columns=name2)
df
```

```python
def NWMatrix(df,match=+1,mismatch=0,gap=-1):
    for i in range(1,m):
        for j in range(1,n):
            if(i==m-1):
                v_score=df.iloc[i,j-1]
                h_score=df.iloc[i-1,j]+gap
            elif(j==n-1):
                v_score=df.iloc[i,j-1]+gap
                h_score=df.iloc[i-1,j]
            else:
                v_score=df.iloc[i,j-1]+gap
                h_score=df.iloc[i-1,j]+gap
            if(df.index[i]==df.columns[j]):
                d_score=df.iloc[i-1,j-1]+match
            else:
                d_score=df.iloc[i-1,j-1]+mismatch
            df.iloc[i,j]=max(v_score,h_score,d_score)   
    return df    
```

```python
NWMatrix=NWMatrix(df,match=+1,mismatch=0,gap=-1)
NWMatrix
```

```python
def BackTracking(NWMatrix,m,n,match=+1,mismatch=0,gap=-1):
    align1=""
    align2=""
    current_score=NWMatrix.iloc[m-1,n-1]
    i=m-1
    j=n-1
    while i>0 and j>0:
            current_score=NWMatrix.iloc[i,j]
            v_score=NWMatrix.iloc[i-1,j]
            h_score=NWMatrix.iloc[i,j-1]
            d_score=NWMatrix.iloc[i-1,j-1]
            if(current_score==v_score+gap):
                align1+=NWMatrix.index[i]
                align2+='-'
                i=i-1
            elif(current_score==h_score+gap):
                align1+='-'
                align2+=NWMatrix.columns[j]
                j=j-1
            else:
                align1+=NWMatrix.index[i]
                align2+=NWMatrix.columns[j]
                i-=1
                j-=1
   
    return align1[::-1],align2[::-1]        
```

```python
BackTracking(NWMatrix,m,n,match=+1,mismatch=0,gap=-1)
```
